## Preprocessing

```python

# importing packages
import pandas as pd
import os
import pandas as pd
from PyPDF2 import PdfReader
```

```python
# Path to the directory containing the PDF files
pdf_dir = r"G:\Codes 2024\Capstone_Final\data"
```

```python
# Initialize an empty list to store extracted text and pdf names
data = []
all_pdf_text = ""  # Initialize an empty string to store all PDF text

# Iterate over PDF files in the directory
for pdf_file in os.listdir(pdf_dir):
    if pdf_file.endswith('.pdf'):
        pdf_path = os.path.join(pdf_dir, pdf_file)
        pdf_name = os.path.splitext(pdf_file)[0]
        
        # Extract text from PDF file
        pdf_text = ''
        with open(pdf_path, 'rb') as pdf:
            pdf_reader = PdfReader(pdf)
            for page_num in range(len(pdf_reader.pages)):
                page = pdf_reader.pages[page_num]
                pdf_text += page.extract_text()
        
        # Append PDF text to the string containing all PDF text
        all_pdf_text += pdf_text
        data.append({"Text": pdf_text, "PDF": pdf_name})

    df = pd.DataFrame(data)
# Print the concatenated PDF text
print(all_pdf_text)
```

```python
df.to_csv('data.csv')
```

## Data Cleaning and Noise Reduction

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# for dwnload models
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Sample text with emojis, hashtags, and other characters
# text = "I love coding! 😊 #PythonProgramming is fun! 🐍✨ Let’s clean some text 🧹"
text = all_pdf_text

# Tokenization
tokens = word_tokenize(text)

# Remove Noise
cleaned_tokens = [re.sub(r'[^\w\s]', '', token) for token in tokens]

# Normalization (convert to lowercase)
cleaned_tokens = [token.lower() for token in cleaned_tokens]

# Remove Stopwords
stop_words = set(stopwords.words('english'))
cleaned_tokens = [token for token in cleaned_tokens if token not in stop_words]

# Lemmatization
lemmatizer = WordNetLemmatizer()
cleaned_tokens = [lemmatizer.lemmatize(token) for token in cleaned_tokens]

# Convert the cleaned tokens back to a string
cleaned_text = ' '.join(cleaned_tokens)
cleaned_text_v1 = cleaned_text
print(cleaned_text_v1)
```

```python
from langchain.text_splitter import CharacterTextSplitter
```

```python
chunk_size = 256
text_spliter = CharacterTextSplitter( # type: ignore
    # separate by stop line
    separator  =" ",  
    # maximum characters per chunk
    chunk_size = chunk_size , 
    # how many characters to overlap chunks
    chunk_overlap = int(chunk_size / 10), 
    # function to calculate the length of a chunk
    length_function = len    
)
# split the text into chunks
chunks = text_spliter.split_text(cleaned_text_v1)
print(chunks)
```

```python
import matplotlib.pyplot as plt
from transformers import AutoTokenizer
```

```python
def plot_doc_tokens(text, EMD_Model):
    tokenizer = AutoTokenizer.from_pretrained(EMD_Model)
    lengths = [len(tokenizer.encode(text))]
    fig = pd.Series(lengths).hist()
    plt.show()
```

```python
plot_doc_tokens(chunks, EMD_Model='sentence-transformers/all-MiniLM-L6-v2')
```
